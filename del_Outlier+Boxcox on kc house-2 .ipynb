```python
import numpy as np
def out_lier_remove(df1):
    for i in range(df1.shape[1]):
        m=df1.iloc[:,i]
        mean=np.mean(m)
        std=np.std(m)
        for j in m:
            z=(j-mean)/std
            if(z>3) or (z<-3):
                df1=df1[df1.iloc[:,i]!=j]
    return df1
```

```python
df1_updated=out_lier_remove(df1)
```

```python
data1=df1_updated.to_csv('clean_kc_house.csv',index=False)
```
