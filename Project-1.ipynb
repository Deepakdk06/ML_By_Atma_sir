```python
import numpy as np
def out_lier_remove(data3):
    for i in range(data3.shape[1]):
        m=data3.iloc[:,i]
        mean=np.mean(m)
        std=np.std(m)
        for j in m:
            z=(j-mean)/std
            if(z>3) or (z<-3):
                data3=data3[data3.iloc[:,i]!=j]
    return data3
```

```python
data3_updated=out_lier_remove(data3)
```
