```python
def Cl(data):
    
    for i in range(len(data.columns)):
        y=data.iloc[:,i]
        out=[]
        mean=np.mean(y)
        std=np.std(y)
        for j in y:
            z=(j-mean)/std
            if(z>3):
                out.append(j)
        for m in out:
            data=data[data.iloc[:,i]!=m]
    return data
        
```
